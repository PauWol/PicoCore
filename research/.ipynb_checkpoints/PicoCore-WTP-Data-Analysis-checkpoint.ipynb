# Voltage and Temperature data analysis
This document deals on evaluating data gathered at night the 17.11.2025 with the PicoCore Weather Test Probe (WTP).

## Dependencies
The dependencies used:

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

## Constants
These are the constants used to evaluate and calculate:

```python
T = 600
REF_TIME = 20:35:15
V_max = 5.3
V_nominal = 5.2
V_min = 3.6 
P_inflection = 0.2
CSV_DATA_FILE = "17-11-2025-WTP-Data.csv"
```

## The Data
Now we load the CSV-File in order to evaluate the data.

```python
df = pd.read_csv(CSV_DATA_FILE)
df.head(10)
```

As we can see `time` is in *ms* since system-boot,then `temperature` in *CÂ°* and `relative humidity` in *%* and the current `voltage` measurement.The time interval $T$ is 10-min in seconds.

> Note: Only 10 samples are displayed out of 59.

## Plotting
In order to see correlations and make data extraction and interpretation easier we plot them.

```python
plt.figure(figsize=(10, 5))  # Width x Height in inches

# Line plots
plt.plot(range(len(df['volt'])), df['volt'], label='Voltage')

# Labels and title
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title('Battery Voltage over Time')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Try to find common column names (adjust list if your CSV uses other names)
time_candidates = ['time', 'timestamp', 'datetime', 'ts']
volt_candidates = ['volt', 'voltage', 'v']
temp_candidates = ['temp', 'temperature', 't']

def find_column(cands):
    for c in cands:
        if c in df.columns:
            return c
    return None

time_col = find_column(time_candidates)
volt_col = find_column(volt_candidates)
temp_col = find_column(temp_candidates)

if volt_col is None:
    raise ValueError(f"No voltage column found. Tried: {volt_candidates}")

# Use sample index if no time column
if time_col:
    try:
        x = pd.to_datetime(df[time_col])
        x_is_time = True
    except Exception:
        x = df[time_col]
        x_is_time = False
else:
    x = list(range(len(df)))
    x_is_time = False

voltage = df[volt_col].astype(float)

# Temperature optional
temp = df[temp_col].astype(float) if temp_col else None

# --------- Plot: big voltage + small temp below ----------
fig, (ax_v, ax_t) = plt.subplots(
    nrows=2, ncols=1,
    gridspec_kw={'height_ratios': [3, 1]},  # voltage plot 3x taller than temp
    figsize=(12, 6),
    sharex=True
)

# Big, easy-to-read voltage plot
ax_v.plot(x, voltage, linewidth=2.8, marker='o', markersize=4)
ax_v.set_title('Voltage over Time (main)', fontsize=16, weight='bold')
ax_v.set_ylabel('Voltage (V)', fontsize=14)
ax_v.grid(True, linestyle='--', alpha=0.5)
ax_v.tick_params(axis='both', which='major', labelsize=12)

# Set a tight y-range for readability if values are known to be ~4.7-5.2 V
# Comment out or adjust the next line if you want automatic scaling:
ax_v.set_ylim(4.6, 5.3)

# Temperature subplot (if available)
if temp is not None:
    ax_t.plot(x, temp, linewidth=1.2, marker='s', markersize=3)
    ax_t.set_ylabel('Temp (Â°C)', fontsize=12)
    ax_t.set_xlabel('Time' if time_col else 'Sample index', fontsize=12)
    ax_t.grid(True, linestyle=':', alpha=0.4)
    ax_t.tick_params(axis='both', which='major', labelsize=10)
else:
    # If no temperature, remove the bottom axes and use only the big voltage plot
    fig.delaxes(ax_t)
    ax_v.set_xlabel('Time' if time_col else 'Sample index', fontsize=12)

# Improve x-axis formatting when using datetimes
if x_is_time:
    fig.autofmt_xdate(rotation=25, ha='right')

plt.tight_layout()
plt.show()
```
