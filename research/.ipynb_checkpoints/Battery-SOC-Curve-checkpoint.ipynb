# Battery state of charge (SOC) and its Discharge Curve
This document deals on creating an average NiMH-Battery discharge curve.

Then there are some changing values tied to the point of requested `SOC`:

## Dependencies:

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

## Graph and Functions
This part deals with the definition of a curve fitting function and a way to plot it:

```python
# Choose a model, e.g., polynomial
def model(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

def model_gradient(x, a, b, c, d):
    return 3*a*x**2 + 2*b*x + c

p_capacity = np.array([1,0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1,0]) # Capacity in percentage
x_fit = np.linspace(0, 1, 100)

def fit_curve(model,voltage):
    global p_capacity
    params, _ = curve_fit(model, p_capacity, voltage)
    return params

def gen_fitted(model,voltage):
    global x_fit 
    params = fit_curve(model,voltage)
    y_fit = model(x_fit,*params)
    return x_fit , y_fit 
    
# Plot
def plot(x,y,voltage,title=None):
    global p_capacity , x_fit 
    plt.plot(p_capacity, voltage, 'o', label='Data')
    plt.plot(x_fit, y, '-', label='Fitted curve')
    plt.xlabel('Capacity (%)')
    plt.ylabel('Voltage (V)')
    if title:
        plt.title(title)
    plt.legend()
    plt.show()
```

## Discharge Curves
In theory a fitting discharge curve could serve as a efficient way to determine the soc fast and somewhat reliable.
The down side of this approach is the need for such a curve with a good accuracy.
The goal of the following is to develop a simple way to normalize the data give and use it to create such a curve to then use it for the soc estimation with an aimed error rate of 20% to 30%.

First data is gathered to then normalize it use it to determine the average curve.

Small sample dataset (20-24CÂ°):

```python

# 1(1.6-0.8):
c_one = np.array([1,1.13,1.15,1.23,1.25,1.27,1.28,1.28,1.29,1.33,1.4])
c_one_fitted = gen_fitted(model,c_one)
plot(*c_one_fitted,c_one)
```

```python
# 2(1.4-1):
c_two = np.array([1,1.15,1.18,1.19,1.20,1.21,1.22,1.22,1.23,1.25,1.33])
c_two_fitted = gen_fitted(model,c_two)
plot(*c_two_fitted,c_two)
```

```python
# 3(12-7.3):
c_three = np.array([7.3,9,9.3,9.5,9.7,9.8,9.9,9.95,10,10.2,11.5])
c_three_fitted = gen_fitted(model,c_three)
plot(*c_three_fitted,c_three)
```

## Averaging the Graphs
In the following the curves are averaged to get a mean curve.As seen below it does not provide the result we need, hence the different input datasets have different scales.

```python
# Average curve
y_fit_avg = (c_one_fitted[1] + c_two_fitted[1] + c_three_fitted[1]) / 3

# Plot all together
plt.plot(*c_one_fitted, label='Fitted 1', linestyle='--')
plt.plot(*c_two_fitted, label='Fitted 2', linestyle='--')
plt.plot(*c_three_fitted, label='Fitted 2', linestyle='--')
plt.plot(x_fit, y_fit_avg, label='Average Fit', color='black', linewidth=2)
plt.xlabel('Capacity (%)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.show()
```

## Normalization (Short Explanation)

Normalization is a process that **scales data to a common range**, usually between 0 and 1. This ensures that datasets with different magnitudes (e.g., voltage ranges like 1â1.4 V vs. 7â11 V) can be compared or averaged fairly.

---

#### Formula:

For any data array ( x ):

$$
\text{normalized}(x_i) = \frac{x_i - \min(x)}{\max(x) - \min(x)}
$$

---

#### Example:

If ( x = [2, 4, 6] ):

$$
\min(x) = 2, \quad \max(x) = 6
$$

Then:

$$
\text{normalized}(x) = \left[ \frac{2-2}{6-2}, \frac{4-2}{6-2}, \frac{6-2}{6-2} \right] = [0, 0.5, 1]
$$

---

#### Purpose:

* Removes unit and scale bias.
* Allows fair averaging across different voltage or gradient ranges.
* Keeps all data within a consistent 0â1 interval.

### Implementation

```python
def normalize(arr):
    """Scale data between 0 and 1 for fair averaging."""
    return (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
```

## Normalizing and Averaging


### Example Comparison of Fitted and Average curve after normalization

```python
y_c_one_fitted_normalized = normalize(c_one_fitted[1])
y_c_two_fitted_normalized = normalize(c_two_fitted[1])
y_c_three_fitted_normalized = normalize(c_three_fitted[1])

# Average curve
y_fit_avg = (y_c_one_fitted_normalized + y_c_two_fitted_normalized + y_c_three_fitted_normalized) / 3

# Plot all together
plt.figure(figsize=(10, 6))
plt.plot(x_fit,y_c_one_fitted_normalized, label='Fitted 1', linestyle='--')
plt.plot(x_fit, y_c_two_fitted_normalized,label='Fitted 2', linestyle='--')
plt.plot(x_fit,y_c_three_fitted_normalized, label='Fitted 2', linestyle='--')
plt.plot(x_fit, y_fit_avg, label='Average Fit', color='black', linewidth=2)
plt.xlabel('Capacity (%)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.show()
```

### Extracting useful data for later usage
This section focuses on extracting additional parameters that are useful for subsequent calculations, particularly those related to estimating discharge behavior and remaining operating time.

- $P_{inflection}$ Defines the fractional increase above the cutoff voltage that marks the end of the nominal discharge range.It represents how much higher the voltage is at the transition point compared to the cutoff voltage (e.g., 10 % â 0.10).

- $V_{end}$ Defines the End of Nominal Range Voltage, i.e. the upper boundary of the final, non-linear discharge region of the battery.Beyond this point, the voltage begins to drop rapidly toward $V_{cutoff}$.

The relationship between both parameters is given by:

$$
V_{end} = V_{cutoff} \times (1+ P_{inflection})
$$


This derived voltage $V_{end}$ is later used in time estimation and discharge modeling formulas, such as calculating the remaining time in the nominal range before reaching the non-linear part of the discharge curve.

The visual evaluation of the above average graph estimates $P_{inflection}$ at about 20% â 0.2 as it is an acceptable point being at x = 0.8 with enough error margin.
This meaning after these last 20% voltage is dropping fast,thus making 20% of it nearly unusable.
